```python
df_countrycode=pd.read_excel("Country-Code.xlsx")
```

```python
#merge this two dataframes
final_df=pd.merge(df,df_countrycode,on='Country Code',how="left")
```

```python
final_df.head()
```

```python
#tocheck dtypes
final_df.dtypes
```

```python
country_name=final_df['Country'].value_counts()
#final_df.Country.value_counts()
```

```python
country_name=final_df['Country'].value_counts().index

country_value=final_df.Country.value_counts().values
```

```python
country_name
```

```python
#top three countries use zomato
plt.pie(country_value[:3],labels=country_name[:3])
```

```python
#top three countries use zomato in prcentage
plt.pie(country_value[:3],labels=country_name[:3],autopct="%1.2f%%")
```

observation:the majority of transaction or activity zomato does is from India following by USA and UK

```python
#grouping 3 columns and reseting the index and canging column name 0 to rating count
ratings=final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:"ratings count"})
```

```python
ratings
```

##observation
  1.when the reating is between 4.5-4.9----> Excellent
  2.when the reating is between 4.0-4.4----> Very good
  3.when the reating is between 3.5-3.9----> Good
  4.when the reating is between 3.0-3.4----> Average
  5.when the reating is between 1.8-2.4----> Poor


```python
#plotting a barplot in seaborn
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sn.barplot(x='Aggregate rating',y='ratings count',data=ratings)
```

```python
matplotlib.rcParams['figure.figsize']=(12,6)
sn.barplot(x='Aggregate rating',y='ratings count',hue='Rating color',data=ratings,palette=['blue','red','orange','yellow','green','green'])
```

observation:
    1.not ratter is very high
    2.most ratings are in2.5-3.5

```python
#countplot
sn.countplot(x='Rating color',data=ratings)
```
